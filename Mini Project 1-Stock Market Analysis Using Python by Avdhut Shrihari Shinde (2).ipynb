```python
    data['Trend'] = data['Daily_Returns']
    data.tail()
    data['Trend']= np.zeros(data['Daily_Returns'].count())
    data['Trend'] = data['Daily_Returns'].apply(lambda x:trend(x))
 
```

```python
data.tail()
```

##  new column ‘Trend’ whose values are based on the 'Daily Return'. 

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return 'Slight or No change'
    elif x > 0.5 and x <= 1:
        return 'Slight Positive'
    elif x > -1 and x <= -0.5:
        return 'Slight Negative'
    elif x > 1 and x <= 3:
        return 'Positive'
    elif x > -3 and x <= -1:
        return 'Negative'
    elif x > 3 and x <= 7:
        return 'Among top gainers'
    elif x > -7 and x <= -3:
        return 'Among top losers'
    elif x > 7:
        return 'Bull run'
    elif x <= -7:
        return 'Bear drop'


data.tail()
```

```python
data.dropna(axis = 0, inplace = True)
data.head()
```

## Visualize trend frequency through a Pie Chart.

```python
fig = plt.figure(dpi = 80)
pie_data = data.groupby('Trend')
pie_label = sorted([i for i in data.loc[:, 'Trend'].unique()])
plt.pie(pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

## What was the correlation between the daily returns of different stocks?

```python
mask = np.zeros_like(corr_data)
mask[np.triu_indices_from(mask)] = True
seaborn.heatmap(corr_data, cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2.5)
plt.yticks(rotation=0) 
plt.xticks(rotation=90) 
plt.show()
```
